```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
```

```python
(train_data, train_labels), (test_data, test_labels) = fashion_mnist.load_data()
```

```python
import matplotlib.pyplot as plt
```

```python
class_names = [
    "T-shirt/top",
    "Trouser/pants",
    "Pullover shirt",
    "Dress",
    "Coat",
    "Sandal",
    "Shirt",
    "Sneaker",
    "Bag",
    "Ankle boot"
    ]
```

```python
import random

fig, ax = plt.subplots(2, 5, figsize=(10, 4))
ax = ax.flatten()
for i in range(10):
    index = random.choice(range(len(train_data)))
    ax[i].imshow(train_data[index], cmap=plt.cm.binary)
    ax[i].set_title(f'{index}.{class_names[train_labels[index]]}')
    ax[i].axis(False)

plt.tight_layout()
plt.show()
```

```python
import random

fig, ax = plt.subplots(2, 5, figsize=(10, 4))
ax = ax.flatten()
for i in range(10):
    index = random.choice(range(len(test_data)))
    ax[i].imshow(test_data[index], cmap=plt.cm.binary)
    ax[i].set_title(f'{index}.{class_names[test_labels[index]]}')
    ax[i].axis(False)

plt.tight_layout()
plt.show()
```

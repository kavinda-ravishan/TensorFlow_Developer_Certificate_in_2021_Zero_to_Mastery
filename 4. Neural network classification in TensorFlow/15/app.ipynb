# Importing libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt

from IPython import display
```

```python
display.set_matplotlib_formats('svg')
```

```python
tf.__version__
```

```python
A = tf.cast(tf.range(-10, 10), tf.float32)
A
```

```python
plt.plot(A, A, 'bo:')
plt.show()
```

```python
def sigmoid(x):

    return 1 / (1 + tf.exp(-x))
```

```python
plt.plot(A, sigmoid(A), 'yo-', markersize=10)
plt.plot(A, tf.keras.activations.sigmoid(A), 'bv:')
plt.title('Sigmoid function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid()
plt.show()
```

```python
def relu(x):

    return tf.maximum(0, x)
```

```python
plt.plot(A, relu(A), 'yo-', markersize=10)
plt.plot(A, tf.keras.activations.relu(A), 'bv:')
plt.title('ReLu function')
plt.xlabel('x')
plt.ylabel('ReLu(x)')
plt.grid()
plt.show()
```

```python
def linear(x):
    return x
```

```python
plt.plot(A, linear(A), 'yo-', markersize=10)
plt.plot(A, tf.keras.activations.linear(A), 'bv:')
plt.title('Linear function')
plt.xlabel('x')
plt.ylabel('Linear(x)')
plt.grid()
plt.show()
```
